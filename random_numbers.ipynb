```python
#write a while loop to perform the same function
#define latitudes as a function and create an empty list
def latitudes(size):
    latitudes = []
    #set variable to zero
    x = 0
    #write the while loop for the function
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x +=1
    return latitudes
#call the function with the size of 1500 and timeit
%timeit latitudes(1500)
```

```python
#write a for loop to do the same thing
randlats = []
x = 0
for x in range (0,1500):
    rand_lat = random.randint(-90, 90) + random.random()
    randlats.append(rand_lat)
    x +=1
%timeit print(randlats)
```
