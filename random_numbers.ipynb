```python
#write a while loop to perform the same function
#define latitudes as a function and create an empty list
def latitudes(size):
    latitudes = []
    #set variable to zero
    x = 0
    #write the while loop for the function
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x +=1
    return latitudes
#call the function with the size of 1500 and timeit
%timeit latitudes(1500)
```

```python
#imort linear regression from the SciPy stats module
from scipy.stats import linregress
```

```python
# Create an equal number of latitudes and temperatures.
lats = [42.5, 43.9, 8.1, 36.8, 79.9, 69.1, 25.7, 15.3, 12.7, 64.5]
temps = [80.5, 75.3, 90.9, 90.0, 40.4, 62.3, 85.4, 79.6, 72.5, 72.0]
```

```python
#perform linear regression. lingregress(x,y)
(slope, intercept, r_value, p_value, std_err) = linregress(lats, temps)
#get the equation of the line
line_eq = 'y = ' + str(round(slope,2)) + 'x +' +str(round(intercept,2))
print(line_eq)
print(f'The p-value is : {p_value:.3f}')
```

```python
#calculate the regrtession line 'y values' from the slope and intercept (y = x*m +b for each value in the y list)
regress_values = [(lat * slope +intercept) for lat in lats]
```

```python
#import matplotlib
import matplotlib.pyplot as plt
#create a scatte plot of the x and y values
plt.scatter(lats, temps)
#plot the regression line with the x-values and the y coordinates (solution to linear equation from above) based on the intercept and slope
plt.plot(lats, regress_values, 'r')
#annotate the text for the lie equation and add its coordinates.
#(what to print, (where to print), size, color)
plt.annotate(line_eq, (10,40), fontsize = 15, color = 'red')
plt.xlabel('Latitude')
plt.ylabel('Temp')
plt.show()
```
